```python
# Part 1: Setup and Data Loading from Wine.csv
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# --- Data Loading ---
# Load your local Wine.csv file
try:
    df = pd.read_csv('Wine.csv')
except FileNotFoundError:
    print("Error: 'Wine.csv' not found. Please ensure the file is in the correct directory.")
    raise

print(f"Dataset Loaded. Original shape: {df.shape}")
print(df.head())

# Define the target column based on your dataset structure
WINE_TYPE_COLUMN = 'Customer_Segment' 

# Separate features (X) from the target variable (y)
# Drop the Customer_Segment column to get only the chemical measurements

X = df.drop(WINE_TYPE_COLUMN, axis=1)
pca = PCA(n_components=2)
xpca = pca.fit_transform(X)
pca_df = pd.DataFrame(xpca)
print(pca_df)

```
